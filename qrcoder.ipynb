```python
from google.colab import files
from PIL import Image
import io
from pyzbar.pyzbar import decode

# Step 3: Upload a PNG QR code image
print("Please upload a PNG image containing a QR code:")
uploaded = files.upload()

# Step 4: Decode the QR code
for filename in uploaded.keys():
    if filename.lower().endswith('.png'):
        image = Image.open(io.BytesIO(uploaded[filename]))

        # Try decoding QR codes
        decoded_objects = decode(image)

        if decoded_objects:
            print("\n✅ QR Code decoded successfully:")
            for obj in decoded_objects:
                print(f"- Type: {obj.type}")
                print(f"- Data: {obj.data.decode('utf-8')}")
        else:
            print("❌ No QR code detected or unable to decode.")
    else:
        print("Only PNG files are supported.")
```
